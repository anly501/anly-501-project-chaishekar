```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
#warning
import warnings
warnings.filterwarnings('ignore')
```

```python
##import clean data
clean_state_data = pd.read_csv('../../data/modified-data/cleaned_state_crime_record_data.csv')
clean_state_data.head()
```

```python
#crime rate by year
figsize=(20, 6)
sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette("RdPu")
sns.lineplot("Year","Property_Crime_Rate",data = clean_state_data, label= 'Property Crime')
sns.lineplot("Year","Violent_Crime_Rate",data = clean_state_data, label= 'Violent Crime')
plt.title('Crime Rate by Year')
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.legend(loc='upper right',);
```

```python
#density plot with legend
fig, ax = plt.subplots(1,2, figsize=(20, 5))
sns.set_style("whitegrid")
sns.set_context("talk")
sns.color_palette("GnBu_d")
sns.kdeplot(clean_state_data['Property_Burglary_Rate'], shade=True, label= 'Burglary Crime', ax=ax[0])
sns.kdeplot(clean_state_data['Property_Larceny_Rate'], shade=True, label= 'Larceny Crime', ax=ax[0])
sns.kdeplot(clean_state_data['Property_Motor_Rate'], shade=True, label= 'Motor Crime',  ax=ax[0])
ax[0].set_title('Property Crime Rate Density Plot')
ax[0].set_xlabel('Crime Rate')
ax[0].set_ylabel('Density')
ax[0].legend(loc='upper right',)
sns.kdeplot(clean_state_data['Violent_Assault_Rate'], shade=True, label= 'Assault', ax=ax[1])
sns.kdeplot(clean_state_data['Violent_Murder_Rate'], shade=True, label= 'Murder', ax=ax[1])
sns.kdeplot(clean_state_data['Violent_Rape_Rate'], shade=True, label= 'Rape', ax=ax[1])
sns.kdeplot(clean_state_data['Violent_Robbery_Rate'], shade=True, label= 'Robbery', ax=ax[1])
ax[1].set_title('Violent Crime Rate Density Plot')
ax[1].set_xlabel('Crime Rate')
ax[1].set_ylabel('Density')
ax[1].legend(loc='upper right',)
plt.show()
```
